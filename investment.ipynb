```
import matplotlib.pyplot as plt # here is the library to plot the data

currentage= 20 # how old you are now
retirementage = 67 # how old you plan to be when you retire

giftamount = 1000 # how much money you've been gifted
interestrate = 0.06 # what is your expected annual return on your investment

balance = [] # setting up an array to hold each year's balance
balance.append(giftamount) # seeding the array with the first value

index = 0 # counter for the loop
while index <= (retirementage - currentage): # run the loop until you retire
  balance.append(balance[index]*(1 + interestrate)) # add next value
  index = index +1 # increment the counter

plt.plot(balance) # plot the data
plt.show() # show the plot
```
